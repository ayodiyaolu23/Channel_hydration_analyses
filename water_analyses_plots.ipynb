```python
import numpy as np
import pandas as pd
import matplotlib.ticker as ticker
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df40 = pd.read_csv('...../1_water-contact-lpa-acyl.dat',sep="\s+", header = None, skiprows = 2)
df41 = pd.read_csv('...../2_water-contact-lpa-acyl.dat',sep="\s+", header = None, skiprows = 2)
```

```python
df40['SMA_200'] = df40.iloc[:,1].rolling(window=200).mean()
df41['SMA_200'] = df41.iloc[:,1].rolling(window=200).mean()
```

```python
fig, ax = plt.subplots(figsize=(7, 5))
ax = plt.subplot(111)
tick_spacing = 1000
ax.xaxis.set_major_locator(ticker.MultipleLocator(tick_spacing))
plt.plot(df40[1],color='white')
plt.plot(df41[1],color='white')
ax.set_xlim(0,5000)
ax.set_ylim(0,30)
ax.set_xlabel(r"Time (ns)", fontsize = 30, fontname="Arial")
ax.set_ylabel(r"Water contacts", fontsize = 30, fontname="Arial");
ax.set_xticklabels(['','0', '100', '200','300', '400','500'], fontsize= 25, fontname="Arial")
ax.tick_params(axis='both', which='major', labelsize=20)
plt.tight_layout()
plt.plot(df40['SMA_200'], color= 'blue', linewidth= 3.5)
plt.plot(df41['SMA_200'], color= 'red', linewidth= 3.5)
x_coordinates = [0, 5000]
plt.legend(bbox_to_anchor=(1,1),loc='upper left', ncol=1, prop={'size':12})
plt.savefig('....../lpa_acyl_water_contact.tiff',dpi=200, bbox_inches="tight")
# display plot
plt.show()
```
